### Intro To Beautiful Soup

## Data from Wikipedia

```python
# Libraries
import requests
import pandas as pd
from bs4 import  BeautifulSoup
from io import StringIO
#import html5lib
import re
```

```python
# Request HTML page
try:
    # Get the page
    url = "https://en.wikipedia.org/wiki/List_of_mammals_of_Germany"
    response = requests.get(url)
    response.raise_for_status()  # Check if for request works else will raise an error
except requests.exceptions.RequestException as e:
    print(f"Error fetching the page: {e}") # to display the error
print(response)
```

```python
soup = BeautifulSoup(response.text, 'html') # Parse the HTML content
#soup = BeautifulSoup(response.text, "html5lib")
print(BeautifulSoup.__module__)  # Check the module being used
```

```python

tag = soup.a # find the first h2 tag
print(type(tag)) # to check the type of the tag
print(tag.name) # this name can be changed if you want to change the tag
print(tag.attrs)
soup.name # the document is not saved anywhere but is referenced in this way
```

```python
soup.find_all('h2') # find all h2 tags
```

```python
head_tag = soup.title # find the title tag
head_tag.contents # .contents gives the children of the tag
order_tag = soup.h2
order_tag.contents

tags = soup.find_all(text = re.compile("Order: +[A-Z]")) # find all h2 tags

orders = []
for tag in tags :
    orders.append(tag.string) # get the string of the tag
print(orders) # to check the orders
```

```python
species = []
species_tags = soup.find_all("i") # find all h2 tags
for tag in species_tags :
    species.append(tag.string) # get the string of the tag
print(species) # to check the species
```

```python
tags = soup.find_all("i")
for tag in tags:
    for child in tag.children:
        print(child) # to check the children of the tag
```

```python
# Finding parent items
header_tag = soup.p
#header_tag.parent


section_tag = soup.div
#section_tag.parent # this gives the parent of the div tag

# it might be cleaner to use a for loop
for parent in header_tag.parents:
    print(parent.name) # this gives the name of the parent tag
    # this will give all the parents of the header tag
    # it will go up to the top of the document
    # and print all the names of the parents
    # this is a generator so it will not print all at once
    # but one by one
```

```python
# In addition to higher level tags and lower levels some tags are at the same level,
# these are called siblings
# I cannot find an example in the wikipedia page
```

```python
tag_nav = soup.nav
for tag in tag_nav.descendants:
    print(tag) # this will find all the tags in the nav tag
    # this will give all the children of the nav tag
    # this is a generator so it will not print all at once
    # but one by one
```
